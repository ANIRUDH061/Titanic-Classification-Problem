# TITANIC SURVIVAL PREDICTION

## Problem Statement
The goal of this project is to predict whether a passenger survived or not based on features like age, gender, ticket class, etc.

```python
## Importing Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Loading the Dataset
df = sns.load_dataset("titanic")
print(df.head())
```

```python
## Defining Features and Target Variable
feature = df[['pclass','sex','age','sibsp','parch','fare']].copy()
target = df['survived']
```

```python
print(feature['sex'].unique())
```

```python
## Data Cleaning
```

```python
feature['sex'] = feature['sex'].map({'male':0, 'female':1})
```

```python
feature['age'] = feature['age'].fillna(feature['age'].mean())
```

```python
## Model Training
```

```python
from sklearn.model_selection import train_test_split
```

```python
## Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pre = model.predict(X_test)
```

```python
## Model Evaluation
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(Y_test,Y_pre))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(Y_test, Y_pre))
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(Y_test, Y_pre)
```

## Conclusion
In this project, we built a machine learning model to predict Titanic passenger survival. 
After data cleaning and feature engineering, the Logistic Regression model was trained and evaluated. 
The model achieved a decent accuracy and showed that features like gender, age, and passenger class play an important role in survival prediction.
